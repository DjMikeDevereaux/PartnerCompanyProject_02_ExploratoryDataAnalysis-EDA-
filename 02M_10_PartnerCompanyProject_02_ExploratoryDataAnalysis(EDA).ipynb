```python
import pandas as pd
df_fact = pd.read_csv("FACT_SALES.csv")
df_fact
```

```python
df_cal = pd.read_excel("DIM_CALENDAR.xlsx")
df_cal
```

```python
df_categ = pd.read_csv("DIM_CATEGORY.csv")
df_categ
```

```python
df_prod = pd.read_excel("DIM_PRODUCT.xlsx")
df_prod
```

```python
df_segm = pd.read_excel("DIM_PRODUCT.xlsx")
df_segm
```

```python
merge_FactProd = pd.merge(df_fact, df_prod, left_on = "ITEM_CODE", right_on = "ITEM")
print(merge_FactProd)
```

```python
pd.merge(df_segm, df_categ, left_on = "CATEGORY", right_on = "ID_CATEGORY")
```

```python
# Distribucion de ventas por region
# La distribucion de las ventas tiene una distribucion similar en las diferentes regiones


import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.boxplot(data=merge_FactProd , x='TOTAL_UNIT_SALES', y='REGION')
plt.title('Distribución de Unidades vendidas por Region')
plt.xticks(rotation=45)
plt.show()
```

```python
fig.savefig("DistribucionVentasRegion.png")
```

```python
# Relación entre Ventas Totales y Unidades vendidas
# Las marcas mas vendidas (color verde) tienen la mayor cantidad de ventas y por lo tanto la mayor cantidad de ingresos
# Los productos con menos cantidad de ventas son los representados en colores rojos

sns.scatterplot(data=merge_FactProd, x='TOTAL_VALUE_SALES', y='TOTAL_UNIT_SALES', hue='BRAND')
plt.title('Relación entre Ventas Totales y Unidades vendidas')
plt.show()
```

```python
fig.savefig("RelaciónVentasTot_UnidadesVendidas.png")
```

```python
# Ventas por region
# La region que mas ingresos genera por ventas es laa 2, despues, muy cerca la 5 y seguidas de la 3, 1 y 4

ventas_region = merge_FactProd.groupby('REGION')['TOTAL_VALUE_SALES'].sum().sort_values()

ventas_region.plot(kind='barh', figsize=(10,6), title='Ventas por región')
plt.xlabel('Ventas')
plt.show()
```

```python
fig.savefig("VentasRegion_barras.png")
```

```python
# Outliers por producto
# Los productos mas representativos son Vanish, Clorox, Cloralex y Blancatel

sns.boxplot(data=merge_FactProd, x='BRAND', y='TOTAL_VALUE_SALES')
plt.title('Outliers en ventas por producto')
plt.xticks(rotation=90)
#plt.show()#
```

```python
fig.savefig("VentasPorPdcto.png")
```

```python
# Ventas por semana 2022 - 2023
# En las ultimas semans del 2023 se observa un incremento en ventas en relacion al promedio 2022 - 2023

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(df_fact["WEEK"], df_fact["TOTAL_VALUE_SALES"])
plt.xticks(rotation=90)
plt.show()
```

```python
fig.savefig("VentasSemanales.png")
```

```python
merge_FactCal = pd.merge(df_fact, df_cal, left_on = "WEEK", right_on = "WEEK")
# Ventas por print(merge_FactCal)
```

```python
# Ventas en ingresos vs tiempo

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(merge_FactCal["DATE"], merge_FactCal["TOTAL_VALUE_SALES"])
plt.xticks(rotation=90)
plt.show()
```

```python
# Ventas en unidades vs tiempo

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(merge_FactCal["DATE"], merge_FactCal["TOTAL_UNIT_SALES"])
plt.xticks(rotation=90)
plt.show()
```

```python
# Ventas 2023
# Se puede observar una tendencia en ventas a finales del año 2023

import matplotlib.pyplot as plt
plt.style.use("ggplot")
fechas = merge_FactCal[merge_FactCal["DATE"] >= "2023-01-01"]

fig, ax = plt.subplots()
ax.plot(fechas["DATE"], fechas["TOTAL_VALUE_SALES"])

plt.xlabel("Datos mensuales; 2023")
plt.ylabel("VENTAS")
plt.xticks(rotation=90)
plt.show()
```

```python
fig.savefig("Ventas2023.png")
```
